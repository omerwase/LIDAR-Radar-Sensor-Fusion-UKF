```python
import pandas as pd

col_names=['Radar NIS']
with open('./radar_nis.txt') as f:
    radar_NIS = pd.read_table(f, sep='\t', header=None, names=col_names, lineterminator='\n')
    
col_names=['Laser NIS']
with open('./laser_nis.txt') as f:
    laser_NIS = pd.read_table(f, sep='\t', header=None, names=col_names, lineterminator='\n')
```

```python
radar_NIS[0:10]
```

```python
laser_NIS[0:10]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.plot(laser_NIS['Laser NIS'], '-o')
plt.axhline(y=5.991, color='r', linestyle='-')
plt.ylabel('NIS')
plt.title('Laser NIS (95% line)')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.plot(radar_NIS['Radar NIS'], '-o')
plt.axhline(y=7.815, color='r', linestyle='-')
plt.ylabel('NIS')
plt.title('Radar NIS (95% line)')
plt.show()
```
